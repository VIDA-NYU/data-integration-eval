```python
import os
import pandas as pd
import json
```

### GDC

```python
base_path = '/Users/pena/nyu-code/data-integration-eval/data/gdc/target-tables/gdc_unique_columns_concat_values.csv'
df = pd.read_csv(base_path, low_memory=False)

df
```

```python
unique_columns = {}

for column in df.columns:
    value_counts = df[column].value_counts()
    if len(value_counts) > 100:
        value_counts = value_counts.head(50)
    unique_columns[column] = value_counts.index.tolist()
unique_columns
```

```python
file_name = 'gdc_columns.json'
with open(file_name, 'w') as f:
    json.dump(unique_columns, f, indent=2)
    print(f"Unique columns saved to {file_name}")
```
