```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = 'colposition/schema_matching_precolposition_gdc_results_20240725153030.csv'

df = pd.read_csv(file_path)

df = df[df['Matcher'] == 'ContrastiveLearning']
df
```

```python
gdc_desc_path = '../data/gdc/gdc_schema_in_gt.csv'
gdc_desc_df = pd.read_csv(gdc_desc_path)
# gdc_desc_df
```

```python
merged_df = df.merge(gdc_desc_df, left_on='Target Column',
                     right_on='column name')
merged_df
```

```python
df_specified = merged_df[merged_df['column values'] != 'Not specified']
df_specified
```

```python
df_nonspecified = merged_df[merged_df['column values'] == 'Not specified']
df_nonspecified
```

```python

gdc_data_path = '../data/gdc/target-tables/gdc_table.csv'
gdc_data_df = pd.read_csv(gdc_data_path)


hard_cases_df = df[(df['Target Column Index'] == -1) |
                   (df['Target Column Index'] > 10)]


for index, row in hard_cases_df.iterrows():
    source_file = row['Source File']
    source_column = row['Source Column']
    target_column = row['Target Column']
    target_index = row['Target Column Index']

    source_data_path = '../data/gdc/source-tables/'+source_file

    df_source = pd.read_csv(source_data_path)

    source_column_values = df_source[source_column].copy()
    target_column_values = gdc_data_df[target_column].copy()

    source_column_values.dropna(inplace=True)
    target_column_values.dropna(inplace=True)
    source_column_values.drop_duplicates(inplace=True)
    target_column_values.drop_duplicates(inplace=True)
    source_column_values.sort_values(inplace=True)
    target_column_values.sort_values(inplace=True)

    sided_vals = pd.concat([source_column_values, target_column_values], axis=1)
    print('Position the match was found:', target_index, source_file)
    print(sided_vals)

   
```
