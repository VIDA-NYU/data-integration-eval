```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Set a better color palette for seaborn
sns.set_palette("Set2")

# Set a better font size for seaborn
sns.set(font_scale=1.2)
# Set the style to white
sns.set_style("white")

# Read the CSV file
file = 'gdc_recallAtK_matching_results_20240711174605.csv'
df = pd.read_csv(file)

# Rest of the code
```

```python
# Melt the DataFrame to long format for seaborn boxplot
df_melted = df.melt(id_vars=["Matcher", "Filename"], 
                    value_vars=["RecallAtK1", "RecallAtK5", "RecallAtK10", "RecallAtK20"],
                    var_name="RecallAtK", value_name="Recall")

# Create the seaborn boxplot
plt.figure(figsize=(14, 7))
sns.boxplot(x="RecallAtK", y="Recall", hue="Matcher", data=df_melted)
plt.title("Recall at K Boxplot for Different Matchers")
plt.show()
```
